```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
classes=["AI","ML","DS","PHP"]
class1_student=[12,23,23,45]
class2_student=[16,27,20,40]
class3_student=[15,29,22,35]

ML_student_age=np.random.randint(18,46,(100))
Py_student_age=np.random.randint(10,38,(100))

classes=["AI","ML","DS","Python"]
class1_student=[56,45,25,70]
```

```python
plt.subplot(3,2,1)

day=[1,2,3,4,5]
delhi_Temp=[22,8,25,30,35]

mumbai_Temp=[13,18,29,37,10]

style.use("ggplot")
plt.plot(day,delhi_Temp,"go--",linewidth=3,markersize=8,label="delhi_Temp Line")
plt.plot(day,mumbai_Temp,"yo-.",linewidth=3,markersize=8,label="mumbai_Temp Line")
plt.axis([0,6,0,40])
plt.xlabel("Day",fontsize=16)
plt.ylabel("Temperature in degree",fontsize=16)
plt.title("Delhi Temperatue",fontsize=18)
plt.grid(color="k",linestyle="-",linewidth=1)
plt.legend(loc=4)
plt.show()



plt.subplot(3,2,2)

explode=[0.3,0,0.2,0]
textprops={"fontsize":18}
colors=['g','r','y','m']
wedgeprops= {"linewidth":0.5,"width":1,"edgecolor":"k"}
plt.pie(class1_student,labels=classes,explode=explode,colors=colors,radius=2,textprops=textprops,autopct="%0.2f%%"
        ,wedgeprops=wedgeprops)
plt.show()


plt.subplot(3,2,3)
from matplotlib import style
style.use("ggplot")
bins=[10,15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ML_student_age,Py_student_age],bins,rwidth=0.8,color=['m','y'],orientation='vertical',label=['Ml_student','Py_student'])
plt.title("ML_student_age")
plt.xlabel("Student_age")
plt.ylabel("Number of student")
plt.grid(color='k',linestyle="-")
plt.legend(loc=4)
plt.show()

plt.subplot(3,2,4)
style.use("ggplot")
class_index=np.arange(len(classes))
width=0.2
plt.figure(figsize=(16,9))
plt.bar(class_index,class1_student,color="y",width=0.2,label="class_1_student")
plt.bar(class_index+width,class2_student,color="g",width=0.2,label="class_2_student")
plt.bar(class_index+width+width,class3_student,color="r",width=0.2,label="class_3_student")

plt.grid(color="k",linestyle="-",linewidth="2")
plt.title("Tecnical Courses",fontsize=25)
plt.xlabel("Classes",fontsize=20)
plt.ylabel("No. of students",fontsize=20)
plt.legend(loc=4)
plt.xticks(class_index+width,classes)
plt.show()

plt.subplot(3,2,5)
plt.subplot(3,2,6)

```
