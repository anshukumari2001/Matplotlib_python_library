# Line Plot

```python
import matplotlib.pyplot as plt
```

```python
month=['Dec','Jan','Feb','Mar','Apr']
Temperature=[22,8,25,30,35]
```

```python
plt.plot(month,Temperature)
plt.xlabel("Month")
plt.ylabel("Temperature in degree")
plt.title("Delhi Temperatue")
plt.show()
```

```python
day=[1,2,3,4,5]
Temperature=[22,8,25,30,35]
```

```python
plt.plot(day,Temperature)
plt.axis([0,5,0,40])
plt.xlabel("Day")
plt.ylabel("Temperature in degree")
plt.title("Delhi Temperatue")
plt.show()
```

```python
plt.plot(month,Temperature,color="r", marker="o", linestyle="--",linewidth=3,markersize=8)
plt.xlabel("Month",fontsize=16)
plt.ylabel("Temperature in degree",fontsize=16)
plt.title("Delhi Temperatue",fontsize=18)
plt.legend(["Temp Line"],loc=4)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.plot(month,Temperature,"go--",linewidth=3,markersize=8,label="Temp Line")
plt.xlabel("Month",fontsize=16)
plt.ylabel("Temperature in degree",fontsize=16)
plt.title("Delhi Temperatue",fontsize=18)
plt.grid(color="k",linestyle="-",linewidth=1)
plt.legend(loc=4)
plt.show()
```

```python
day=[1,2,3,4,5]
delhi_Temp=[22,8,25,30,35]

mumbai_Temp=[13,18,29,37,10]

style.use("ggplot")
plt.plot(day,delhi_Temp,"go--",linewidth=3,markersize=8,label="delhi_Temp Line")
plt.plot(day,mumbai_Temp,"yo-.",linewidth=3,markersize=8,label="mumbai_Temp Line")
plt.axis([0,6,0,40])
plt.xlabel("Day",fontsize=16)
plt.ylabel("Temperature in degree",fontsize=16)
plt.title("Delhi Temperatue",fontsize=18)
plt.grid(color="k",linestyle="-",linewidth=1)
plt.legend(loc=4)
plt.show()
```
